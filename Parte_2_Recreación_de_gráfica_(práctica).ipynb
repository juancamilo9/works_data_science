```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
filename = '/content/drive/MyDrive/mortgages.csv'
df =  pd.read_csv(filename)
df.head()
```

```python
filter_30_year = df['Mortgage Name'] == '30 Year'

df[filter_30_year]
```

```python
filter_interest_rate_3 = df['Interest Rate'] == 0.03
df[filter_interest_rate_3]
```

```python
#DataFrame para 30 year y 0.03%
df_30_03_year = df[filter_30_year & filter_interest_rate_3]
df_30_03_year
```

```python
# DataFrame para 0.05%
df_30_05_year = df[filter_30_year & ~filter_interest_rate_3]
df_30_05_year
```

```python

plt.style.use('classic')
plt.figure(dpi=100)
plt.plot(df_30_03_year['Month'],df_30_03_year['Interest Paid'].cumsum(),c='b',label='30 Year 3%')
plt.plot(df_30_05_year['Month'],df_30_05_year['Interest Paid'].cumsum(),c='k',label='30 Year 5%')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Month',  fontsize=15)
plt.ylabel('Dollar', fontsize=15)
plt.title('Cumulative Interest Paid')
plt.legend(loc=(1.02,0))
plt.grid()

```
