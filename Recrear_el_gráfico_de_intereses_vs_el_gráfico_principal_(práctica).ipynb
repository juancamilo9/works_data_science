```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
filename = '/content/drive/MyDrive/mortgages.csv'
df =  pd.read_csv(filename)
df.head()
```

**Creen un filtro para que el nombre de la hipoteca sea de “30 años”.  Usen len() para encontrar cuántas filas corresponden a 30 años.**

```python
filter_30_year = df['Mortgage Name'] == '30 Year'

df[filter_30_year]
```

**Creen un filtro para una tasa de interés del 3 % (observen cómo está escrito en los datos).**

```python
filter_interest_rate_3 = df['Interest Rate'] == 0.03
df[filter_interest_rate_3]
```

**Creen un filtro para una tasa de interés del 3 % (observen cómo está escrito en los datos).**

```python
#DataFrame para 30 year y 0.03%
df_30_03_year = df[filter_30_year & filter_interest_rate_3]
df_30_03_year
```

```python
# DataFrame para 0.05%
df_30_05_year = df[filter_30_year & ~filter_interest_rate_3]
df_30_05_year
```

**Puesto que queremos graficar la suma acumulativa del interés pagado, podemos usar la función .cumsum().   ¿Cuáles son los tres primeros valores en la salida para la suma acumulativa de la tasa del 3 %?**

```python
df_30_03_year.cumsum().head(3)
```

**Puesto que queremos graficar la suma acumulativa del interés pagado, podemos usar la función .cumsum().   ¿Cuáles son los tres primeros valores en la salida para la suma acumulativa de la tasa del 3 %?**

```python
plt.plot(df_30_03_year['Month'],df_30_03_year['Interest Paid'].cumsum(),c='b')
plt.plot(df_30_05_year['Month'],df_30_05_year['Interest Paid'].cumsum(),c='k')
plt.show()
```
