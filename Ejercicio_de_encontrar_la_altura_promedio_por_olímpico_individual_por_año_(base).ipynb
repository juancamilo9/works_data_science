#  Creen filtros para encontrar:

**¿Cuántos atletas que participaron en el deporte "Tug-Of-War" (tira y afloja) pertenecían en el equipo “Sweden” (sueco) en el conjunto de datos de muestra?**

```python
# filto prar 'Tug-Of-War'
filter_TW = df['Sport'] == 'Tug-Of-War'
# filtro para 'Sweden'
filter_sweden = df['Team'] == 'Sweden'
# cuento con respecto al ID del atleta
print("Numero Total de entradas:" , df[filter_TW & filter_sweden]['ID'].count())
df[filter_TW & filter_sweden]

```

# b. ¿Cuántas medallas de oro se le entregaron a los atletas que tenían más de 40 años?

```python
filter_age = df['Age'] > 40
filter_medal = df['Medal'] == 'Gold'
# contamos con respecto al ID del atleta
print("Total de atletas mayores de 40 años que ganaron medalla de oro:" , df[filter_medal & filter_age]['ID'].count())
df[filter_medal & filter_age]
```
